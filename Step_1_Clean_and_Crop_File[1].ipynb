```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the IndiaAgricultureCropProduction.csv dataset into a Pandas DataFrame
main_dataset=pd.read_csv("/content/drive/MyDrive/CSCT MASTER PROJECT/IndiaAgricultureCropProduction.csv", sep = ",", low_memory= False)
```

```python
# Extract the year part from the 'Year' column
main_dataset['Year'] = main_dataset['Year'].str[:4].astype(int)
```

```python
# Count the number of null values in each column using isnull() and sum()
null_counts = main_dataset.isnull().sum()
```

```python
# Displaying the null counts for each column
print(null_counts)
```

```python
# Dropping rows with null values using the dropna() method
clean_dataset = main_dataset.dropna()
```

```python
# Counting the number of null values in each column using the isnull() and sum() methods
null_counts = clean_dataset.isnull().sum()
# Displaying the null counts for each column
print(null_counts)
```

```python
# Checking for duplicate rows
duplicate_rows = clean_dataset.duplicated()
```

```python
# Counting the number of duplicate rows
num_duplicate_rows = sum(duplicate_rows)
```

```python
# Printing the result
if num_duplicate_rows > 0:
    print("There are", num_duplicate_rows, "duplicate rows in the dataset.")
else:
    print("There are no duplicate rows in the dataset.")
```
