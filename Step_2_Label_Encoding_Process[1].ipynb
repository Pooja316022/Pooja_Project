```python
## Selecting non-numeric data columns for further processing
agri_obj = df_new.select_dtypes(object) #non numeric data
```

```python
# Selecting numeric data columns for further analysis
agri_num = df_new.select_dtypes(["int64","float64"]) # numeric data
```

```python
# Creating a copy of the original DataFrame for data encoding
df_encoded = df_new.copy()
```

```python
# Initialize the LabelEncoder
le = LabelEncoder()
```

```python
# Encoding categorical variables using Label Encoding
# Iterating through each column containing non-numeric data in 'agri_obj'
# and transforming it into numerical format in the 'df_encoded' DataFrame
for col in agri_obj.columns:
    df_encoded[col] = le.fit_transform(agri_obj[col])
```

```python
# Encoding categorical variables using Label Encoding
# Applying Label Encoding to each column containing non-numeric data in 'agri_obj'
# and updating the corresponding columns in 'df_encoded'
df_encoded[agri_obj.columns] = agri_obj.apply(lambda col: le.fit_transform(col))
```

```python
# Display the first few rows of the encoded DataFrame 'df_encoded' to inspect the encoded categorical data
print(df_encoded.head())
```

```python
# Save the encoded DataFrame 'df_encoded' to a CSV file named 'df_encoded.csv' without including the index
df_encoded.to_csv('/content/drive/MyDrive/CSCT MASTER PROJECT/df_encoded.csv', index=False)
```
